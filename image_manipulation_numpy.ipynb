```python
dog_image=np.array(plt.imread("dog.jpeg"))
dog_image.shape
```

```python

plt.imshow(dog_image)
```

Rotating the image


```python

plt.imshow(dog_image[::-1,: : ,:])
```

croping the image

```python
plt.imshow(dog_image[20:250,120:420,:])
```

Blur image


```python
plt.imshow(dog_image[::50,::50,:])
```

```python

image_contrast=np.where(dog_image>150,255,0)
plt.imshow(image_contrast)
```

```python
image_negative=255-dog_image
plt.imshow(image_negative)
```

```python

image_copy=dog_image.copy()
image_copy[:,:,(1,2)]=0
plt.imshow(image_copy)
```
